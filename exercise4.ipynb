```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("question4.csv")
df.head()
```

```python
#to find the null values - 
df.isna().sum()
```

```python
#Question 1 - 
#Ignore the missing values in the dataset - 
df_ignore = df.dropna()
df_ignore.head()
```

```python
#filling nan values manually - 
df1 = df.copy()
```

```python
df1.loc[df1["Age"].isnull(), "Age"] = 28
df1.loc[df1["Salary"].isnull(), "Salary"] = 50000  
df1.loc[df1["Department"].isnull(), "Department"] = "Sales"
```

```python
df1.head()
```

```python
#Filling NaN values with Global constant - 
df_global = df.fillna("UNK")
df_global.head()
```

```python
#filling values with central tendency - 
df_central = df.copy()
df_central["Age"] = df["Age"].fillna(df["Age"].median()) #median so as to avoid decimal answers in age.
df_central["Salary"] = df["Salary"].fillna(df["Salary"].mean())
df_central["Department"] = df["Department"].fillna(df["Department"].mode()[0])
df_central.head()
```

```python
df_probalistic = df.copy()
df_probalistic["Age"] = df_probalistic["Age"].interpolate() 
df_probalistic["Salary"] = df_probalistic["Salary"].interpolate()
df_probalistic["Department"] = df_probalistic["Department"].ffill() 
df_probalistic.head()
```
